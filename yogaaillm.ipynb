### **Introduction to Artificial Intelligence (AI)**  
Artificial Intelligence (AI) is the science of making computers think and act like humans. It helps machines understand, learn, and solve problems just like we do. AI is used in everyday life, such as voice assistants (like Siri or Alexa), medical diagnosis, and even self-driving cars.  

In yoga and health sciences, AI can help analyze posture, suggest personalized exercises, and even detect stress levels using facial expressions or voice tone.

---

### **What is a Large Language Model (LLM)?**  
A **Large Language Model (LLM)** is a type of AI that understands and generates human language. It is trained on massive amounts of text and can answer questions, summarize information, and even hold conversations.  

For example, an LLM can be used to:  
‚úÖ Translate yoga texts from Sanskrit to English  
‚úÖ Generate personalized meditation scripts  
‚úÖ Provide answers to health-related queries  

LLMs power AI chatbots and virtual assistants that can interact with users in a natural way.

---

### **What is Retrieval-Augmented Generation (RAG)?**  
**Retrieval-Augmented Generation (RAG)** is a technique that improves AI-generated answers by fetching the most relevant information from external sources before generating a response.  

üîπ Imagine you ask a chatbot, "What are the benefits of Surya Namaskar?"  
üîπ Instead of just guessing, the AI will first **retrieve** information from a yoga textbook or research paper.  
üîπ Then, it will **generate** a well-informed answer based on that data.  

This ensures that AI responses are more **accurate, reliable, and up-to-date**.

##Why is Chunking Performed?

Chunking is the process of dividing a long text into smaller, manageable parts (chunks). It is performed for several reasons:

*     **Efficient Retrieval**: Instead of searching the entire document, the system can quickly fetch relevant chunks, improving response time.
*     **Better Context Understanding**: AI models, including LLMs, have a limit on how much text they can process at once. Chunking ensures that each query gets relevant and focused context.
*     **Improved Accuracy**: Smaller chunks prevent AI from retrieving too much irrelevant text, leading to more precise answers.
*     **Memory Optimization**: Handling large texts in smaller blocks reduces memory consumption and speeds up processing.
*     **Enhances RAG Performance**: A well-chunked document ensures that the retrieval system fetches the most relevant text before generating responses.

##Optimal Chunk Size for the NCERT Yoga Textbook
For building a RAG-based chatbot, the ideal chunk size depends on:

*     **The complexity of the content**
*     **The structure of the textbook (sections, headings, subheadings)**
*     **The token limit of the LLM used for generation (e.g., OpenAI's GPT models or LLaMA models)**

Based on these factors:
*     ‚úÖ Recommended Chunk Size: 250‚Äì500 words (or ~1000‚Äì2000 tokens)
*     ‚úÖ Overlap Between Chunks: 50‚Äì100 words (to maintain context continuity)

**üîπ Why this range?**

*     **Less than 250 words**: May lose important context, leading to incomplete answers.
*     **More than 500 words**: Might retrieve too much irrelevant text, reducing retrieval efficiency.
*     **Overlap of 50‚Äì100 words**: Helps maintain coherence between chunks so that answers are not disjointed.

````python
from pypdf import PdfReader
chunk = []

def overlap_addpdfchunks(name, overlap=None):
	reader = PdfReader(name) #Open the pdf, create a pdf object and store in variable 'reader'
	page_num = 0 #Keep count of page number. Initially 0
	char_count = 0 #Keep count of the global characters on the whole pdf
	#Go through each page of the pdf and break each page to chunks
	for page in reader.pages:
		page_num += 1 #Everytime a page gets opened, increment page no by 1
		temp_chunk = "" #Temporary varialbe to store the chuck under consideration until it is appended. Once a chunk is appended, re-initialize it to null ""
		#Go through every page character by character and keep appending each character to temp_chunk until (1) chararacter count reaches 300 or (2) the page ends
		overlap_chars = ""
		overlap_appended = False
		for charac in page.extract_text():
			if not overlap_appended:
				temp_chunk += overlap_chars
				overlap_appended = True
			temp_chunk += charac #adding each character to temp_chunk
			char_count += 1 #increment global character counter by 1
			if len(temp_chunk) == 300: #if chuck size is reached
				chunk.append(temp_chunk) #append chunk to chunklist
				overlap_chars = temp_chunk[-overlap:] # Overlap from the back
				overlap_appended = False
				temp_chunk = "" #reinitialize the temp chunk storage variable
		chunk.append(temp_chunk) #for last chunk in the page if char count is less than 300, add to chunklist
		print("Number of pages completed: ", page_num)
	print("Number of character in the yoga.pdf: ", char_count)
	return char_count

overlap_num_chars = overlap_addpdfchunks("yoga.pdf", 100)
````

### **Why is Vector Embedding Required? (With Respect to Yoga & Health Science)**  

In simple terms, **vector embedding** converts text into numbers (vectors) so that a computer can understand and compare it efficiently. This is crucial for making an **AI-powered Yoga Q&A chatbot** because:  

1. **Understanding Yoga Concepts Like a Human** üßò‚Äç‚ôÇÔ∏è  
   - Traditional search methods rely on exact words. But in yoga, similar concepts can be expressed differently (e.g., "deep breathing" vs. "pranayama").  
   - Vector embeddings help the AI **understand meaning, not just words**, making it more effective in answering questions.  

2. **Fast & Intelligent Searching** üîç  
   - If a student asks, "What are the benefits of Trikonasana?" the AI doesn‚Äôt search for exact words but **finds related information** even if the phrase is different in the book.  
   - This is **better than keyword-based searches**, which might miss relevant information.  

3. **Grouping Similar Concepts** üìö  
   - The AI can recognize that "Meditation" and "Mindfulness" are related, even if they don‚Äôt use the same words.  
   - This helps in providing **holistic and context-aware answers** in yoga and health.

ChromaDB is a **vector database** designed for **storing and retrieving text efficiently using embeddings**. It is used because:  

1. **Fast Retrieval for Large Textbooks** üìñ

````python
import requests
import json
import os

def get_chunks_from_chroma(query):
  result = yogachunks.query(query_texts=[query], n_results=10)["documents"]
  return result

def send_to_llm(query):
  from google.colab import userdata
  GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

  URL = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent?key={GOOGLE_API_KEY}"

  headers = {
    "Content-Type": "application/json"
  }

  data = {
    "contents": [
        {
            "parts": [
                {"text": """ You are a question answering machine designed only to answer questions Yoga and Helath Science. No other system. If the question is not relevant to Yoga, say that the question is out of context and ask the user to finetune their query for Yoga and Health Science. Also explain with relavanet example why it is out of context. If no system is mentioned assume it means Yoga.

        I am going to give you a question and some relevant chunks as output from a program. You must combine the chunks in a meaningful manner so as to be the answer to the given question. If the chunks do not sufficiently answer the question feel free to use your own knowledge of Yoga to answer. Expalain with a step-by-step running example when applicable
              """},
                {"text": str(get_chunks_from_chroma(query))},
                {"text":str(query)}

            ]
        }
    ]
  }

  response = requests.post(URL, headers=headers, data=json.dumps(data))
  return response.json()["candidates"][0]["content"]["parts"][0]["text"]
````
